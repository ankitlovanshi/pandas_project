```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Sample data
data = {
    'Value': [10, 20, 30, 40, 1000, 30, 40, 50, 60, 70]  # 1000 is an outlier
}
df = pd.DataFrame(data)

# Box plot to visualize outliers
plt.boxplot(df['Value'])
plt.title('Box Plot of Value')
plt.show()

# Using IQR to detect outliers
Q1 = df['Value'].quantile(0.25)
Q3 = df['Value'].quantile(0.75)
IQR = Q3 - Q1

# Lower bound and upper bound
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Identifying outliers
outliers = df[(df['Value'] < lower_bound) | (df['Value'] > upper_bound)]
print("Outliers:")
print(outliers)

# Removing outliers
df_cleaned = df[(df['Value'] >= lower_bound) & (df['Value'] <= upper_bound)]
print("Data without outliers:")
print(df_cleaned)
```

```python
df =  pd.DataFrame({
    'City': ['New York', 'Los Angeles', 'New York', 'Chicago', 'Houston', 
             'Los Angeles', 'Houston', 'Bhopal', 'New York', 'Los Angeles'],
    'Salary': [85000, 90000, 75000, 70000, 65000, 
               80000, 72000, 69000, 85000, 91000],
    'Department': ['Finance', 'Engineering', 'HR', 'Marketing', 'Sales', 
                   'Civil', 'Machenical', 'Marketing', 'Finance', 'Manangement']
    
})
df
```

```python
# 800000 ---> 30
# 60000 ----> 25
# 40000 ----> 20
# 10000 ----> 18
```

```python
df['Age'] = range(10)
df
```

```python
def add_age(col):
    if col > 80000:
        return 30
        
    elif col > 60000:
        return 25
        
    elif col > 40000:
        return 20
    else:
        return 18
        
df['Age'] = df['Salary'].apply(add_age)
df
```
