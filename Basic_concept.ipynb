```python
# Data Transformation
# Applying functions (apply, applymap, map)
# Grouping and aggregation (groupby, agg, transform)
```

```python
data = {
    'NumericalColumn': [1, 2, np.nan, 4],
    'CategoricalColumn': ['A', 'B', np.nan, 'A'],
    'BooleanColumn': [True, np.nan, False, True],
    'StringColumn': ['foo', 'bar', np.nan, 'baz']
}
df = pd.DataFrame(data)
```

```python
df.info()
```

```python
# apply
def fill_na(col):
    if col.dtype == 'object':
        return col.fillna('unknow')
    elif col.dtype == 'int64':
        return col.fillna(0)
    else:
        return col.fillna(0.0)

df = df.apply(fill_na)
        
```

```python
df
```

```python
# applymap
df = pd.DataFrame({
    'A': [1, 2, 3],
    'B': [4, 5, 6]
})

def squre(x):
    return x ** 2

df.applymap(squre)
```

```python
# map
df = pd.DataFrame({
    'A': [1, 2, 3],
    'B': [4, 5, 6]
})

def square_column(x):
    return x ** 2

df['Square of A'] = df['A'].map(square_column)
df

# skip
```

```python
# transform
df = pd.DataFrame({
    'A': [1, 2, 3, 4],
    'B': [10, 20, 30, 40]
})

def square(x):
    return x ** 2

df.transform(square)
```

```python
# Applying a Transformation to Each Group

df = pd.DataFrame({
    'Team': ['A', 'A', 'B', 'B', 'C', 'C'],
    'Points': [10, 20, 15, 25, 10, 30]
})
def square(x):
    return x ** 2

df.groupby('Team')['Points'].sum().transform(square)
```

```python
# Combining and Merging Data:

# Concatenation (concat)
# Merging and joining (merge, join)
```

```python
# Concatenation
df1 = pd.DataFrame({
    'A': [1, 2],
    'B': [3, 4]
})

df2 = pd.DataFrame({
    'A': [5, 6],
    'B': [7, 8]
})
result = pd.concat([df1, df2], axis=0)
result
```

```python
# merge
df1 = pd.DataFrame({
    'key': ['A', 'B', 'C'],
    'value1': [1, 2, 3]
})

df2 = pd.DataFrame({
    'key': ['B', 'C', 'D'],
    'value2': [4, 5, 6]
})
result = pd.merge(df1 ,df2, how='inner', on='key')
result
```
