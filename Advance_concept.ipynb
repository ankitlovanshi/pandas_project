```python
# String manipulation
data = {
    'Name': ['Alice Johnson', 'Bob Smith', 'Charlie Brown', 'David Lee', 'Eva Green'],
    'Email': ['alice.johnson1@example.com', 'bob.smith@example.com', 'charlie.brown@example.com', 'david.lee@example.com', 'eva.green@example.com'],
    'Phone': ['+1-555-1234', '+1-555-5678', '+1-555-8765', '+1-555-4321', '+1-555-3456']
}
df = pd.DataFrame(data)
```

```python
# contains
df[df['Email'].str.contains('example')]
```

```python
# extract
df['Username_extracted'] = df['Email'].str.extract(r'(^[\w\.]+)')
print(df[['Email', 'Username_extracted']])
```
