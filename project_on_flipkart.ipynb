```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
import plotly.express as px
```

```python
# read csv
flip = pd.read_csv('/home/developer/Downloads/flipkart.csv')
```

```python
flip
```

```python
# get info
flip.info()
```

```python
# check null values
flip.isnull().sum()
```

```python
# drop null values
flip.dropna(how='all', inplace=True)
```

```python
flip
```

```python
# check null values
flip.isnull().sum()
```

```python
flip.info()
```

```python
# fill null values
flip.fillna({'retail_price' : flip['retail_price'].mean(), 
            'discounted_price' : flip['discounted_price'].mean(), 
            'brand' : 'Unkown', 'product_specifications' : 'Unkown'}, inplace=True)
```

```python
flip.isnull().sum()
```

```python
flip
```

```python
# Price Distribution:
sns.histplot(flip['retail_price'], bins=15, kde=True)
```

```python
# Top Brands by Number of Products:
df = flip.groupby('brand').size().sort_values(ascending=False).reset_index().head(10)
df=df.rename(columns = {0 : 'total product'})
sns.lineplot(df, x='brand', y='total product')
plt.xticks(rotation='vertical')
```

```python
df
```

```python
flip.columns
```

```python
# Correlation Analysis:
correlation = flip[['retail_price', 'discounted_price', 'product_rating']].corr()
```

```python
# Product Ratings and Discounts:
sns.scatterplot(data=flip, y='discounted_price', x='product_rating')
plt.xticks(rotation='vertical')
plt.show()
```

```python
flip.iloc[0]
```

```python
# Use Plotly for Interactive Visualizations:
fig = px.scatter(flip, y='discounted_price', x='product_rating')
fig.show()
```

```python
df=flip.groupby('product_category_tree').size().reset_index().sort_values(by=0, ascending=False).head(15)
sns.barplot(data=df, x='product_category_tree', y=0)
plt.xticks(rotation='vertical')
plt.title("Most sould out 15 Category")
plt.xlabel('Product Category')
plt.ylabel('Count of product')
plt.show()
```
